### 딥-러닝 과정 Mulit Layer Perceptron(MLP) 

## Simple Keras 모델 생성/학습 - XOR Function

```python
# 1. Numpy 가져오기

import numpy as np

print(np.__version__)
```

```python
# 2. 입력/출력 데이터 만들기
X = np.array([[0,0],[0,1],[1,0],[1,1]])
y = np.array([[0],[1],[1],[0]])

print(X)
print(y)
```

```python
# 3. Keras 패키지 가져오기

from tensorflow import keras
from tensorflow.keras import layers

print(keras.__version__)
```

```python
# 4. MLP 모델 생성
model = keras.Sequential()
model.add(layers.Dense(units=8, input_dim=2))
model.add(layers.Activation('relu'))
model.add(layers.Dense(1))
model.add(layers.Activation('sigmoid'))

model.summary()
```

```python
# 5. Compile - Optimizer, Loss function 설정
sgd = keras.optimizers.SGD(lr = 0.1)
model.compile(loss='binary_crossentropy', 
            optimizer=sgd,
            metrics=['accuracy'])
```

```python
# 6. 학습시키기
model.fit(X,y,
        batch_size=1 ,
        epochs= 500) 
```

```python
# 7. 모델 테스트하기
test_data = np.array([[0,1],[1,1],[0,0],[1,0]]) # 1 0 0 1
result = model.predict(test_data)

print(result)
```

```python
print(model.predict_classes(test_data) > 0.5)
```
