### 딥-러닝 과정 Mulit Layer Perceptron(MLP) 

## Simple Keras 모델 생성/학습 - AND Function

```python
# 1. Numpy 가져오기
import numpy as np

print(np.__version__)
```

```python
# 2. 입력/출력 데이터 만들기
X = np.array([[0,0],[0,1],[1,0],[1,1]])
Y = np.array([[0],[0],[0],[1]])
X,Y
```

```python
# 3. Keras 패키지 가져오기
from tensorflow import keras
from tensorflow.keras import layers

print(keras.__version__)
```

```python
# 4. Perceptron 모델 생성
model = keras.Sequential()

#네트워크 쌓기
model.add(layers.Dense(units=1, input_dim=2))
model.add(layers.Activation('sigmoid'))

model.summary()
```

```python
# 5. Compile - Optimizer, Loss function 설정

sgd = keras.optimizers.SGD(lr=0.1)
model.compile(loss='binary_crossentropy',
                    optimizer='sgd') 
```

```python
# 6. 학습시키기
model.fit(X,Y,
        batch_size=1 ,        # 문제 몇개를 풀고 답을 확인 할 것인가?  문제 푸는갯수에 따라 속도 가달라짐
        epochs= 500)            #반복횟수 
```

```python
# 7. 모델 테스트하기
test_data = np.array([[0,1],[1,1],[0,0],[1,0]]) 
result = model.predict(test_data)

print(result)
```

```python
# 예측
# print(model.predict_classes(test_data)) 구형 코드 

print(model.predict_classes(test_data) > 0.5)
```
